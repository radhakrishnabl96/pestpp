### Reading the results

```python
case1_obs = pd.read_csv('Case1_su_cv_valid.demo/observations.dat',skiprows=190,sep=' ')
case5_obs = pd.read_csv('Case5_su_cv_valid.demo/observations.dat',skiprows=262,sep=' ')
case5_III_obs = pd.read_csv('Case5_III_su_cv_valid.demo/observations.dat',skiprows=220,sep=' ')


case1_obs.head()
```

```python
# Converting seconds to days

case1_obs['time [days]'] = case1_obs['time [s]']/86400
case5_obs['time [days]'] = case5_obs['time [s]']/86400
case5_III_obs['time [days]'] = case5_III_obs['time [s]']/86400
```

```python
### Temperature from degree celcius to Kelvin

depths = [0.04, 0.1, 0.2, 0.4, 0.8, 1.2, 1.6]

for i, depth in enumerate(depths):
    case1_obs[f'Soil temperature at - {depth} m (°C)'] = case1_obs[f'point -{depth} temperature [K]'] - 273.15
    case5_obs[f'Soil temperature at - {depth} m (°C)'] = case5_obs[f'point -{depth} temperature [K]'] - 273.15
    case5_III_obs[f'Soil temperature at - {depth} m (°C)'] = case5_III_obs[f'point -{depth} temperature [K]'] - 273.15


case1_obs.head()
```

```python
### Saturation of liquid to Volumetric Water Content (VWC - %)

depths = [0.04, 0.1, 0.2, 0.4, 0.8, 1.2, 1.6]

depth_peat = 0.3

poro_peat_1 = 0.63840773749824
poro_mineral_1 = 0.48163834735695

poro_peat_2 = 0.42152389790647
poro_mineral_2 = 0.36232061622228

poro_peat_3 = 0.64661880775129
poro_mineral_3 = 0.48321096221786



for i, depth in enumerate(depths):
    if depth < depth_peat:
        case1_obs[f'Soil moisture at - {depth} m (%)'] = case1_obs[f'point -{depth} saturation liquid']*(poro_peat_1*100)
        case5_obs[f'Soil moisture at - {depth} m (%)'] = case5_obs[f'point -{depth} saturation liquid']*(poro_peat_2*100)
        case5_III_obs[f'Soil moisture at - {depth} m (%)'] = case5_III_obs[f'point -{depth} saturation liquid']*(poro_peat_3*100)
    else:
        case1_obs[f'Soil moisture at - {depth} m (%)'] = case1_obs[f'point -{depth} saturation liquid']*(poro_mineral_1*100)
        case5_obs[f'Soil moisture at - {depth} m (%)'] = case5_obs[f'point -{depth} saturation liquid']*(poro_mineral_2*100)
        case5_III_obs[f'Soil moisture at - {depth} m (%)'] = case5_III_obs[f'point -{depth} saturation liquid']*(poro_mineral_3*100)
```

```python
# Defining the dataframe

depths = [0.04, 0.1, 0.2, 0.4, 0.8, 1.2, 1.6]

nse_df_case1 = pd.DataFrame(index=depths,columns=['Temperature','Moisture'])
nse_df_case2 = pd.DataFrame(index=depths,columns=['Temperature','Moisture'])
nse_df_case3 = pd.DataFrame(index=depths,columns=['Temperature','Moisture'])
```

```python
# Defining a Nash-Sutcliffe coefficient
def nse(targets, predictions):
    nse_value = (1-(np.sum((predictions-targets)**2)/np.sum((targets-np.mean(targets))**2)))
    return nse_value
```

```python
# Test
nse(df_obs_AWS[f'Soil temperature at - 1.6 m (°C)'], case5_obs[f'Soil temperature at - 1.6 m (°C)'])
```
