- The outputs from the pest analysis of the four cases is analyzed here. 

The initial conditions remain the same for all cases: I.C : Temperature = -3°C, Hydrostatic head = - 1 m.

    1. Case1_B : (Case1_B_glm_rw_restart2) - Top B.C: Subsurface temperature at 0.0 m; No. of parameters = 20
    2. Case5_I_E : (Case5_I_E_glm2_rw) - SEB (Surface energy balance) without snow - **Monthly** data of incoming_shortwave_radiation, precipitation, air temperature, relative humidity, wind speed; No. of parameters = 26
    3. Case5_II_A : (Case5_II_A_glm_cf_v1_rw) - SEB (Surface energy balance) without snow - **Daily** data of incoming_shortwave_radiation, precipitation, air temperature, relative humidity, wind speed; No. of parameters = 26
    4. Case5_III_C : (Case5_III_C_glm_cf_v1_rw) - SEB (Surface energy balance) **with snow** - **Daily** data of incoming_shortwave_radiation, precipitation, air temperature, relative humidity, wind speed, and snow depth; No. of parameters = 26

Note:
- All of the input files was generated from the jupyter notebooks: {Case_name}_pestpp_fileprep

```python
# Finding the index where we will be encountering the minimum objective function

#itrno_best = obj_iteration['total_phi'].idxmin()

for i, name in enumerate(cf_names):
    # Extracting the best iteration by the least value of the objective function
    locals()[f'itrno_best_{i+1}'] = locals()[f'obj_fn_{i+1}']['total_phi'].idxmin()
    
    # The objective function for the best iteration 
    locals()[f'obj_itr_best_{i+1}'] = locals()[f'obj_fn_{i+1}'].iloc[locals()[f'itrno_best_{i+1}'],:]
    
    # Calculating the contribution of the variable temperature and moisture
    locals()[f'con_var_{i+1}'] = np.array(locals()[f'obj_itr_best_{i+1}'][['temp','mois']])/locals()[f'obj_itr_best_{i+1}']['measurement_phi']*100
    

    
```
